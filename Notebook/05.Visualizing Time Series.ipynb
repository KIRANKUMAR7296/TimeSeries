```python
import pandas as pd
```

```python
df = pd.read_csv('Starbucks.csv', index_col='Date', parse_dates=['Date'])
df.head()
```

```python
df['Close'].plot(figsize=(14,6), title='Closing Price', xlabel='Date', ylabel='Price');
```

```python
ax = df['Close'].plot(figsize=(14,6), title='Closing Price')
ax.autoscale(axis='x', tight=True)
ax.set(xlabel='Date', ylabel='Price');
```

```python
ax = df['Close']['2017-01':'2017-12-31'].plot(figsize=(14,6), title='Closing Price', ls='--')
ax.autoscale(axis='x', tight=True)
ax.set(xlabel='Date', ylabel='Price');
```

Set **xlim** : X Axis Limit.

```python
ax = df['Close'].plot(figsize=(14,6))
ax.autoscale(axis='x', tight=True)
ax.set(title='Closing Price', xlabel='Date', ylabel='Price', xlim=['2017-01-01','2017-12-31']);
```

```python
from matplotlib import dates
```

CODE | MEANING
:---: | :---
**%Y** | Year with **century** : 2021
**%y** | Year without **century** : 21
**%m** | Month : 02
**%B** | Month Full Name : February
**%b** | Month Abbreviation : Feb
**%d** | Day of the month : 07
**%A** | Weekday Full Name : Monday
**%a** | Weekday abbreviation : Mon
**%H** | Hour (24 hour clock) : 13
**%I** | Hour (12 hour clock) : 01
**%p** | AM or PM
**%M** | Minute
**%S** | Second
**%#m** | Month : 2 (Windows)
**%-m** | Month : 2 (Mac)
**#x** | Long Date : Monday, February 02, 2021
**#c** | Long Date & Time : Monday, February 02, 2021 16:05:55


```python
df['Close'].plot(xlim=['2017-01-01','2017-03-01'], ylim=[50,60],figsize=(12,6));
```

# Set Ticks

```python
ax = df['Close'].plot(xlim=['2017-01-01','2017-03-01'], ylim=[50,60],figsize=(12,6))
ax.set(xlabel='')
ax.xaxis.set_major_locator(dates.WeekdayLocator(byweekday=0))
ax.xaxis.set_major_formatter(dates.DateFormatter('%d'))

ax.xaxis.set_minor_locator(dates.MonthLocator())
ax.xaxis.set_minor_formatter(dates.DateFormatter('\n\n%b'))

ax.yaxis.grid(True);
# byweekday = 0 (Weekday starts from Monday)
# x ticks are located or equally spaced by weekdays.
```
